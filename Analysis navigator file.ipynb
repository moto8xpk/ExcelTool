```python
mytree = ET.parse('navigator.xml')
def removeSingleColumn(keyOfCustomField):
    rootMinor=mytree.getroot().find('exportinclusion')
    newRoot=mytree.getroot().find('exportinclusion').getchildren()

    tempOrder=0
    deleted=False

    for index,item in enumerate(newRoot):
        if(item.tag=="field"):
            if(deleted):
                tempOrder+=1
                item.find('order').text=str(tempOrder)

            if(item.attrib.get('name')==keyOfCustomField):
                tempOrder=int(item.find('order').text)-1
                deleted=True
                rootMinor.remove(item)
                print('have removed==========================================================>',keyOfCustomField)
    mytree.write("newxml.xml")
    return True
print(removeSingleColumn('ttconcession.field.applianceFamily'))
```

```python
mytree = ET.parse('navigator.xml')
checkNumber=2
tempOrderUpdated=3
def addNewCustomFieldAfterToNavigator(orderBeforeToUpdate,orderCurrentToUpdate,keyCustomField,issueTypeList):
    updated=False

    homeRoot=mytree.getroot().find('exportinclusion')
    for index,item in enumerate(mytree.getroot().find('exportinclusion').iter('field')):
        if(item.tag=="field"):
            if(item.attrib.get('name')==keyCustomField):
                break;
            if(updated):
                orderCurrentToUpdate+=1
                item.find('order').text=str(orderCurrentToUpdate)
            if(int(item.find('order').text)==orderBeforeToUpdate):
                updated=True
                field=ET.SubElement(homeRoot, 'field')
                field.set('name', keyCustomField)
                issuetypes=ET.SubElement(field, 'issuetypes')
                
                for itemIssueType in issueTypeList:
                    issuetype=ET.SubElement(issuetypes, 'issuetype')
                    issuetype.text=itemIssueType

                order=ET.SubElement(field, 'order')
                order.text=str(orderCurrentToUpdate)
    mytree.write("newxmlv2.xml")
    return True
issueList=['ttconcession.issuetype.default4',
           'ttconcession.issuetype.default6',
           'ttconcession.issuetype.default7']
addNewCustomFieldAfterToNavigator(78,79,"helloFieldUpdate",issueList)
```
